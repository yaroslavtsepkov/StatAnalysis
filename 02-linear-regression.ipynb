```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
```

```python
n = 10000
a, b = 2, -2
sigmasqaure = 1
x = np.linspace(0, 1,num=n)
y = a * x + b + np.random.normal(scale=sigmasqaure**(1/2), size=n)
```

## gen data

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
```

```python
print(f'slope: {slope}')
print(f'intercept: {intercept}')
print(f'R_square: {r_value**2}')
print(f'p_value: {p_value}')
print(f'std_err: {std_err}')
```

```python
fig, ax = plt.subplots(1,1)
ax.scatter(x=x, y=y, edgecolors='white', c='black')
ax.plot(x, slope*x+intercept, c='purple', linewidth=3)
```

## abalone data

```python
abalone = pd.read_csv('data/abalone.data', usecols=['Length','Diameter'])
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(abalone.Diameter, abalone.Length)
print(f'slope: {slope}')
print(f'intercept: {intercept}')
print(f'R_square: {r_value**2}')
print(f'p_value: {p_value}')
print(f'std_err: {std_err}')
```

```python
fig, ax = plt.subplots(1,1)
ax.scatter(x=abalone.Diameter, y=abalone.Length, c='k', edgecolors='white', label='scatter')
ax.plot(x,slope*x+intercept, c='purple', linewidth = 3, label='linear regresion')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('abalone regression model')
ax.legend()
fig.savefig('abalone-regression-model.png', dpi=500)
```
